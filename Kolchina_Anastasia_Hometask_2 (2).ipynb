У данных метрик разный принцп работы. BLEU основан на сравнении с человеческим переводом, который считается эталонным. Для этого каждое предложение сравнивается с набором качественных референсных переводов, после чего полученное качество усредняется по всему тексту. Смысл текста и грамматические ошибки не учитываются, что может являться некоторым недостатком. Однако метрика имеет большое преимущество засчет ее сравнения с человеческим переводом, также она работает достаточно быстро.

BLEURT же является метрикой, обученной на BERT с помощью transfer learning c использованием синтетических данных. В конце концов метрика обучается на WMT человеческих аннотациях. Преимуществом данной метрики является то, что если ее дообучить, она даст хороший результат, максимально приближенный к человеческой оценке. Из недостатков стоит отметить низкую скорость работы (засчет столь сложного алгоритма с применением нейронной сети и дообучением), а также то, что без этого самого дообучения метрика работает хуже, чем BLEU.

Эта модель предсказывает вероятность того, что входное предложение приемлемо (грамматически верно) в русском языке.
```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("RussianNLP/ruRoBERTa-large-rucola")

model = AutoModelForSequenceClassification.from_pretrained("RussianNLP/ruRoBERTa-large-rucola")
```

```python
! pip install torch
```

```python
import torch

if torch.cuda.is_available(): # кладем на гпу
    model = model.cuda()
```

```python
df_gr.head()
```

```python
import torch.nn.functional as F
import numpy as np

device = torch.device('cuda:0')

sort_1 = []
sort_5 = []
sort_8 = []

for i in range(20000):
    if torch.cuda.is_available():
        out_1 = tokenizer(df_gr['trans-1'][i], return_tensors="pt", padding=True).to(device)
        out_5 = tokenizer(df_gr['trans-5'][i], return_tensors="pt", padding=True).to(device)
        out_8 = tokenizer(df_gr['trans-8'][i], return_tensors="pt", padding=True).to(device)

    with torch.no_grad():
        probs_1 = F.softmax(model(**out_1).logits, dim=1)
        probs_5 = F.softmax(model(**out_5).logits, dim=1)
        probs_8 = F.softmax(model(**out_8).logits, dim=1)

    curr = np.array([df_gr['trans-1'][i], df_gr['trans-5'][i], df_gr['trans-8'][i]])
    res = [probs_1[0][0].cpu(), probs_5[0][0].cpu(), probs_8[0][0].cpu()]
    idx = np.argsort(np.array(res), axis=0)
    curr = curr[idx]
    sort_1 += [curr[0]]
    sort_5 += [curr[1]]
    sort_8 += [curr[2]]
```

```python
translations = pd.DataFrame()
translations['sort_1'] = sort_1
translations['sort_5'] = sort_5
translations['sort_8'] = sort_8
```

### **[2 балла]** Обучите NMT модель используя обучающие, валидационные и тестовые данные **строго по назначению**. Можно воспользоваться готовыми фреймворками ([huggingface](https://huggingface.co/), [simpletransformers](https://simpletransformers.ai/)) или пайплайнами (например, [этим](https://github.com/devjwsong/transformer-translator-pytorch)).

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("t5-small")
```

```python
source_lang = "en"
target_lang = "tr"
prefix = "translate English to Turkish: "


def preprocess_function(examples):
    inputs = [prefix + example for example in examples[source_lang]]
    targets = [example for example in examples[target_lang]]
    model_inputs = tokenizer(inputs, text_target=targets, max_length=128, truncation=True)
    return model_inputs
```

```python
train_df.shape
```

```python
val_df.shape
```

```python
from datasets import Dataset

train_dataset = Dataset.from_pandas(train_df[:40000])
```

```python
val_dataset = Dataset.from_pandas(val_df[:10000])
```

```python
test_dataset = Dataset.from_pandas(test_df)
```

```python
tokenized_train = train_dataset.map(preprocess_function, batched=True)
```

```python
tokenized_val = val_dataset.map(preprocess_function, batched=True)
```

```python
tokenized_test = test_dataset.map(preprocess_function, batched=True)
```

```python
from transformers import AutoModelForSeq2SeqLM

model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")
```

```python
from transformers import DataCollatorForSeq2Seq

data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)
```

```python
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer

training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=1,
    fp16=True,
    predict_with_generate=True
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    tokenizer=tokenizer,
    data_collator=data_collator,
)

trainer.train()
```

```python
pred = trainer.predict(tokenized_test)
preds = tokenizer.batch_decode(pred.predictions, skip_special_tokens=True)
```
